## Mobile Apps Data Analysis

```python
from csv import reader

### The Google Play data set ###
opened_file = open('googleplaystore.csv',encoding='utf-8')
read_file = reader(opened_file)
android = list(read_file)
android_header = android[0]
android = android[1:]
```

```python
##Exploring data
def explore_data(dataset, start, end, rows_and_columns=False):
    dataset_slice = dataset[start:end]    
    for row in dataset_slice:
        print(row)
        print('\n') # adds a new (empty) line between rows
        
    if rows_and_columns:
        print('Number of rows:', len(dataset))
        print('Number of columns:', len(dataset[0]))

print(android_header)
print('\n')
explore_data(android, 0, 3, True)
```

```python
#checking duplicate entries
for app in android:
    name = app[0]
    if name == 'Instagram':
        print(app)
```

```python
#checking all duplcates and making list
duplicate_apps = []
unique_apps = []

for app in android:
    name = app[0]
    if name in unique_apps:
        duplicate_apps.append(name)
    else:
        unique_apps.append(name)
    
print('Number of duplicate apps:', len(duplicate_apps))
print('\n')
print('Examples of duplicate apps:', duplicate_apps[:15])
```

```python
#length of our dictionary (of unique apps) should be equal to the difference between the length of our data set and 1,181.
```

```python
reviews_max = {}

for app in android:
    name = app[0]
    n_reviews = float(app[3])
    
    if name in reviews_max and reviews_max[name] < n_reviews:
        reviews_max[name] = n_reviews
        
    elif name not in reviews_max:
        reviews_max[name] = n_reviews
```

```python
# verifying
print('Expected length:', len(android) - 1181)
print('Actual length:', len(reviews_max))
```

```python
# data seggregating
android_clean = []
already_added = []

for app in android:
    name = app[0]
    n_reviews = float(app[3])
    
    if (reviews_max[name] == n_reviews) and (name not in already_added):
        android_clean.append(app)
        already_added.append(name)
```

```python
#verifying
explore_data(android_clean, 0, 3, True)
```

## Removing non-english apps

```python
# Assuming long names having more than 3 non-ASCII as non english and see the result
def is_english(string):
    non_ascii = 0
    
    for character in string:
        if ord(character) > 127:
            non_ascii += 1
    
    if non_ascii > 3:
        return False
    else:
        return True

print(is_english('Docs To Goâ„¢ Free Office Suite'))
print(is_english('Instachat ðŸ˜œ'))
#The hypothesis is not 100% accurate but its an experiment to see
```

```python
#Filtering out non english app
android_english = []

for app in android_clean:
    name = app[0]
    if is_english(name):
        android_english.append(app)
        
explore_data(android_english, 0, 3, True)
```

```python
#seggregating only 'free' apps
android_final = []

for app in android_english:
    price = app[7]
    if price == '0':
        android_final.append(app)
print(len(android_final))
```

## Most Common Apps Genre

```python
def freq_table(dataset, index):
    table = {}
    total = 0
    
    for row in dataset:
        total += 1
        value = row[index]
        if value in table:
            table[value] += 1
        else:
            table[value] = 1
    
    table_percentages = {}
    for key in table:
        percentage = (table[key] / total) * 100
        table_percentages[key] = percentage 
    
    return table_percentages


def display_table(dataset, index):
    table = freq_table(dataset, index)
    table_display = []
    for key in table:
        key_val_as_tuple = (table[key], key)
        table_display.append(key_val_as_tuple)
        
    table_sorted = sorted(table_display, reverse = True)
    for entry in table_sorted:
        print(entry[1], ':', entry[0])
```

```python
# Categorywise frequency
display_table(android_final, 1)
```

```python
#Most popular app by genre
display_table(android_final, 5)
```

```python
## Most popular app by genre
#Avg no of install by category
categories_android = freq_table(android_final, 1)

for category in categories_android:
    total = 0
    len_category = 0
    for app in android_final:
        category_app = app[1]
        if category_app == category:            
            n_installs = app[5]
            n_installs = n_installs.replace(',', '')
            n_installs = n_installs.replace('+', '')
            total += float(n_installs)
            len_category += 1
    avg_n_installs = total / len_category
    print(category, ':', avg_n_installs)
```
